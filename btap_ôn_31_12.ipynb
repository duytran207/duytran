```python
GPA = float(input("GPA = "))
if GPA >= 8.5:
    print("Xuất sắc")
elif 7.0 <= GPA < 8.5:
   print("giỏi")
elif 5.5 <= GPA < 7:
   print("khá")
elif 4.0 <= GPA < 5.5:
   print("trung bình")
else:
  GPA < 4
  print("yếu")
```

```python
n = int(input("n = "))
tong = 0

for i in range(1, n + 1):
  if i % 2 == 0:
    tong += i
print("tổng số chẵn từ 1 tới",n ,"có tổng là :", tong)
```
