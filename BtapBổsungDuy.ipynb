```python
numbers = []
print("Nhập 10 số nguyên:")

for i in range(10):
    x = int(input(f"Số thứ {i+1}: "))
    numbers.append(x)

so_duong = 0
so_am = 0
so_khong = 0

for num in numbers:
    if num > 0:
        so_duong += 1
    elif num < 0:
        so_am += 1
    else:
        so_khong += 1

print("Số dương:", so_duong)
print("Số âm:", so_am)
print("Số bằng 0:", so_khong)
```

```python
n = int(input("Nhập một số từ 1 đến 9: "))

if 1 <= n <= 9:
    print(f"Bảng cửu chương của {n}:")
    for i in range(1, 10):
        print(f"{n} x {i} = {n * i}")
else:
    print("Vui lòng nhập số trong khoảng 1 đến 9.")
```

```python
n = int(input("Nhập n: "))

if n < 2:
    print("Không phải số nguyên tố")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Không phải số nguyên tố")
            break
    else:
        print("Là số nguyên tố")
```
